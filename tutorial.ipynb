### Importing targets from a csv file

For importing objects from a csv file the column indices taken are 'name', 'ra', 'dec' so ensure that the columns are labelled accordingly. The default delimiter is set to a comma however this can be updated by the 'delimiter' kwarg.

```python
#pyviz.obs_plot(objects, observatory id)

pyviz.obs_plot('template_import.csv', 'lasilla')
```

### Other keyword arguments

The date can be specified in either MJD or yyyy-mm-dd format.

```python
#pyviz.obs_plot(objects, observatory id, date = today)

pyviz.obs_plot('template_import.csv', 'lasilla', date = '2022-11-14')
```

The moon angle can be displayed for each object by passing 'moon_angle' as True, and minimum and maximum observing angles can be set with 'min_angle' and 'max_angle' respectively.

```python
#pyviz.obs_plot(objects, observatory id, moon_angle = False, min_angle = None, max_angle = None)

pyviz.obs_plot('template_import.csv', 'lasilla', moon_angle = True, min_angle = 30, max_angle = 80)
```

Pointing restrictions can be introduced by specifying disallowed nautical directions in the list. The sections of the object paths corresponding to these directions shall be emitted from the plot.

```python
#pyviz.obs_plot(objects, observatory id, pointing_restrictions = None)

pyviz.obs_plot('template_import.csv', 'lasilla', pointing_restrictions = ['N', 'NW'])
```

Finally the plot can be saved by specifying a path for the output file with the 'outfile' kwarg.

```python
#pyviz.obs_plot(objects, observatory id, outfile = None)

pyviz.obs_plot('template_import.csv', 'lasilla', outfile = './tutorial.png')
```
